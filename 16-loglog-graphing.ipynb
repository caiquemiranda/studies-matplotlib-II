### Chapter 16: LogLog Graphing

#### LogLog graphing

```python
import matplotlib.pyplot as plt
import numpy as np

A = 30
a = 3.5
x = np.linspace(0.01, 5, 10000)
y = A * x ** a
ax = plt.gca()

plt.plot(x, y, 
         linewidth = 2.5, 
         color = 'navy', 
         label = r'$f(x) = 30 \cdot x^{3.5}$')

plt.legend(loc='upper left')
plt.xlabel(r'x')
plt.ylabel(r'y')

ax.grid(True)

plt.title(r'Normal plot')
plt.show()
plt.clf()

xlog = np.log(x)
ylog = np.log(y)
ax = plt.gca()

plt.plot(xlog, ylog, linewidth=2.5, color='navy', label=r'$f(x) = 3.5\cdot x + \ln(30)$')
plt.legend(loc='best')
plt.xlabel(r'log(x)')
plt.ylabel(r'log(y)')

ax.grid(True)
plt.title(r'Log-Log plot')
plt.show()
plt.clf()
```
